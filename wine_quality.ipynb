```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("/content/drive/MyDrive/winequality-white.csv", sep=";")
columns = data.columns.to_list()
x_colmns = [col for col in columns]
x_colmns.remove("quality")
for con in x_colmns:
    y = np.array([data["quality"].to_list()])
    x = np.array([data[con].to_list()])
    plt.xlabel(con)
    plt.ylabel("quality")
    plt.plot(x[0],y[0])
    print("\n")
    plt.show()
```
